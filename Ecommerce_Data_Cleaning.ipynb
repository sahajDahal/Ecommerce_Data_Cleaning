```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/My Drive/datasets/brazil_Ecommerce.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
# Translating column names from Portuguese to English
df = df.rename(columns = {'Título':'Title','Reais':'price(reais)','Centavos':'cents(reais)','Condição':'condition','Nota':'average_rating',
                     'N_Avaliações':'number_of_reviews','Desconto':'discount_applied','Marca':'brand','Gênero':'gender',
                     'Temporada':'season'})
```

```python
# Removing unnecessary columns
df.drop(columns = ['Unnamed: 0','season'],inplace = True)
```

```python
df.head(3)
```

```python
df.dtypes
```

```python
# Combining price and cents column by adding cents to total price
df['price(reais)'] += (df['cents(reais)']/100)
```

```python
df.drop(columns = 'cents(reais)', inplace = True)
df.head(3)
```

```python
# Removing parenthesis from number_of_reviews column
df['number_of_reviews'] = df['number_of_reviews'].str.strip('()')
df.head(3)
```

```python
# Changing null values in number_of_reviews column to 0
# Changing datatype of number_of_reviews column to integer
df['number_of_reviews'] = df['number_of_reviews'].fillna(0).astype('int64')
df['number_of_reviews'] = df['number_of_reviews'].astype('int64')
df.dtypes
```

```python
df.head(3)
```

```python
# Converting discount_applied column into a float for easier mathematical analysis in the future
df['discount_applied']=df['discount_applied'].str.replace('% OFF','')
df['discount_applied'] = df['discount_applied'].astype('float')
df['discount_applied'] = df['discount_applied'].fillna(0.0)
df['discount_applied'] /= 100
```

```python
# Removing rows with null price data
df.dropna(subset = ['price(reais)'],inplace = True)
df['price(reais)'].isna().sum()
```
